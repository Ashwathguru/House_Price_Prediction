```python
totaldata = totaldata.dropna(subset=['MSZoning','SaleType'])
```

```python
totaldata.head()
```

```python
mean_value = totaldata['LotFrontage'].mean()
totaldata['LotFrontage'].fillna(mean_value, inplace=True)
```

```python
totaldata.HouseStyle.unique()
```

```python
totaldata.HouseStyle.value_counts()
```

```python
def replace_titles(x):
    title=x['HouseStyle']
    if title in ['1Story']:
        return 1
    elif title in ['2Story']:
        return 2
    elif title in ['1.5Fin']:
        return 1.75
    elif title in ['SLvl','SFoyer']:
        return 1.5 
    elif title in ['2.5Fin']:
        return 2.75
    elif title in ['2.5Unf']:
        return 2.5
    elif title in ['1.5Unf']:
        return 1.5
    else:
        return title
totaldata['HouseStyleint']=totaldata.apply(replace_titles, axis=1)
```

```python
totaldata.head()
```

```python
totaldata = totaldata.drop(['HouseStyle'],axis=1)
```

```python
totaldata.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
le.fit(totaldata['MSZoning'])


totaldata['MSZoning'] = le.transform(totaldata['MSZoning'])

le.fit(totaldata['Street'])

totaldata['Street'] = le.transform(totaldata['Street'])


```

```python
le.fit(totaldata['LotShape'])
totaldata['LotShape'] = le.transform(totaldata['LotShape'])

le.fit(totaldata['LandContour'])
totaldata['LandContour'] = le.transform(totaldata['LandContour'])


le.fit(totaldata['LotConfig'])
totaldata['LotConfig'] = le.transform(totaldata['LotConfig'])


le.fit(totaldata['Neighborhood'])
totaldata['Neighborhood'] = le.transform(totaldata['Neighborhood'])

le.fit(totaldata['SaleType'])
totaldata['SaleType'] = le.transform(totaldata['SaleType'])

le.fit(totaldata['SaleCondition'])
totaldata['SaleCondition'] = le.transform(totaldata['SaleCondition'])
```

```python
totaldata.to_csv('clean.csv')
```
